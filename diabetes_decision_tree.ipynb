```
import pandas as pd

dbt = pd.read_csv('/content/drive/MyDrive/diabetes_data.csv')
```

```
dbt.head(50)
```

```
dbt.drop('Pregnancies',axis=1,inplace=True)
dbt.drop('DiabetesPedigreeFunction',axis=1,inplace=True)
```

```
x = dbt[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','Age']]
y = dbt[['Outcome']]
```

```
from sklearn.tree import DecisionTreeClassifier
a=DecisionTreeClassifier()
a.fit(x,y)
```

```
gula = int(input('Masukkan nilai gula darah : '))
darah = int(input('Masukkan tekanan darah : '))
kulit = int(input('Masukkan nilai ketebalan kulit :'))
insulin = int(input('Masukkan kadar insulin : '))
bmi = float(input('Masukkan nilai Body Mass Index : '))
umur = int(input('Masukkan umur : '))

hsl = a.predict([[gula,darah,kulit,insulin,bmi,umur]])
if hsl :
  print("Pasien diprediksi diabetes")
else :
  print("Pasien diprediksi tidak diabetes")
```

```
a.score(x,y)
```
